### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
df.info()
df.describe(include="all").T
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
# Identify missing values
missing_counts = df.isna().sum().sort_values(ascending=False)
print(missing_counts.head(20))

# Impute missing values
df["YearsCodePro"] = pd.to_numeric(df["YearsCodePro"], errors="coerce")
df["YearsCodePro"] = df["YearsCodePro"].fillna(df["YearsCodePro"].median())

for col in ["JobSat", "Employment", "RemoteWork"]:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
print("Employment types:\n")
print(df["Employment"].value_counts())

print("\nJob satisfaction values:\n")
print(df["JobSat"].value_counts())

print("\nYearsCodePro statistics:\n")
print(df["YearsCodePro"].describe())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
job_counts = df["JobSat"].value_counts()

plt.figure(figsize=(10, 30))
job_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("Job Satisfaction Distribution")
plt.ylabel("")   # hide y-label
plt.tight_layout()
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
lang_worked = (
    df["LanguageHaveWorkedWith"]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
)

lang_want = (
    df["LanguageWantToWorkWith"]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
)

# top 10 languages in each list
top_n = 10
top_worked = lang_worked.value_counts().head(top_n)
top_want = lang_want.value_counts().head(top_n)
lang_compare = pd.DataFrame({
    "WorkedWith": top_worked,
    "WantToWorkWith": top_want
}).fillna(0)

print(lang_compare)

# bar chart
plt.figure(figsize=(12, 6))
lang_compare.plot(kind="bar", figsize=(12, 6))
plt.title("Top Programming Languages: Worked With vs Want To Work With")
plt.xlabel("Programming Language")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
remote_df = df[["Country", "RemoteWork"]].dropna()

remote_table = pd.crosstab(remote_df["Country"], remote_df["RemoteWork"])
top_countries = remote_df["Country"].value_counts().head(8).index
remote_table = remote_table.loc[top_countries]

# stacked bar chart
remote_table.plot(kind="bar", stacked=True, figsize=(12, 6))
plt.title("Remote Work Preferences by Country")
plt.xlabel("Country")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Remote Work")
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
corr_data = df[['JobSat', 'YearsCodePro']].copy()

corr_data['JobSat'] = pd.to_numeric(corr_data['JobSat'], errors='coerce')
corr_data['YearsCodePro'] = pd.to_numeric(corr_data['YearsCodePro'],errors='coerce')

corr_data = corr_data.dropna(subset=['JobSat', 'YearsCodePro'])

print("Dtypes after cleaning:")
print(corr_data.dtypes)
print()
print("First few rows used for correlation:")
print(corr_data.head())
print()
print("Number of records used:", len(corr_data))

pearson_corr = corr_data['JobSat'].corr(corr_data['YearsCodePro'],method='pearson')
print(f"\nPearson correlation between JobSat and YearsCodePro: {pearson_corr:.3f}")

job_rank = corr_data['JobSat'].rank()
exp_rank = corr_data['YearsCodePro'].rank()
spearman_corr = job_rank.corr(exp_rank)

print(f"Spearman correlation between JobSat and YearsCodePro: {spearman_corr:.3f}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
# cross-tab: EdLevel × Employment
edu_emp = pd.crosstab(df["EdLevel"], df["Employment"])

print(edu_emp.head())

# stacked bar chart
edu_emp.plot(kind="bar", stacked=True, figsize=(14, 7))
plt.title("Employment Type by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Employment", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

### Step 10: Export Cleaned Data


- Save the cleaned dataset to a new CSV file for further use or sharing.


```python
## Write your code here
output_file = "survey_data_cleaned_eda.csv"
df.to_csv(output_file, index=False)
print("Cleaned dataset saved as:", output_file)
```
